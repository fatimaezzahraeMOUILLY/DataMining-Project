```python
Irisdataset.drop('Id', axis = 1, inplace = True)
```

```python
# Codage binaire pour la variable 'Species'
Irisdataset["Species"] = Irisdataset["Species"].map({"Iris-setosa": 0, "Iris-versicolor": 1, "Iris-virginica": 2})
```

```python
# Diagramme de dispersion des caractéristiques
sns.pairplot(Irisdataset, hue="Species")
plt.legend(title='Species', loc='upper right', labels=['Setosa', 'Versicolor', 'Virginica'])
plt.show()
```

```python
# Matrice de corrélation
corr_matrix = Irisdataset.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Matrice de corrélation')
plt.show()
```

```python
import pandas.plotting
from pandas.plotting import andrews_curves
andrews_curves(Irisdataset, "Species")
```

```python
from sklearn.model_selection import train_test_split

# Séparation des caractéristiques et de la cible
X = Irisdataset.drop("Species", axis=1)
y = Irisdataset["Species"]

# Division des données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print(X_train),
print(y_train)
```

```python
from sklearn.linear_model import LogisticRegression

# Initialisation et entraînement du modèle
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Prédiction sur les données de test
y_pred = model.predict(X_test)
```

```python
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score

# Évaluation de la précision
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import confusion_matrix

# Matrice de confusion
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
```

```python
from sklearn.metrics import classification_report

# Rapport de classification
class_report = classification_report(y_test, y_pred)
print("Classification Report:")
print(class_report)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report


# Initialisation et entraînement du modèle d'arbre de décision
decision_tree_model = DecisionTreeClassifier()
decision_tree_model.fit(X_train, y_train)

# Prédiction sur les données de test
y_pred_decision_tree = decision_tree_model.predict(X_test)

# Évaluation du modèle d'arbre de décision
accuracy_decision_tree = accuracy_score(y_test, y_pred_decision_tree)
print("Accuracy (Decision Tree):", accuracy_decision_tree)

# Matrice de confusion
conf_matrix_decision_tree = confusion_matrix(y_test, y_pred_decision_tree)
print("Confusion Matrix (Decision Tree):")
print(conf_matrix_decision_tree)

# Rapport de classification
class_report_decision_tree = classification_report(y_test, y_pred_decision_tree)
print("Classification Report (Decision Tree):")
print(class_report_decision_tree)
```

```python
from sklearn.svm import SVC

# Initialisation et entraînement du modèle SVM
svm_model = SVC()
svm_model.fit(X_train, y_train)

# Prédiction sur les données de test
y_pred_svm = svm_model.predict(X_test)

# Évaluation du modèle SVM
accuracy_svm = accuracy_score(y_test, y_pred_svm)
print("Accuracy (SVM):", accuracy_svm)

# Matrice de confusion
conf_matrix_svm = confusion_matrix(y_test, y_pred_svm)
print("Confusion Matrix (SVM):")
print(conf_matrix_svm)

# Rapport de classification
class_report_svm = classification_report(y_test, y_pred_svm)
print("Classification Report (SVM):")
print(class_report_svm)
```
